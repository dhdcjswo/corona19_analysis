```python
# 중국 내 데이터만 선별 & sum
c = c.loc[c['Country/Region']=='Mainland China'].sum()[4:]
d = d.loc[d['Country/Region']=='Mainland China'].sum()[4:]
c
```

```python
# 2차함수 fitting
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score, mean_absolute_error

y = c.values[:15]
x = np.arange(len(y))

poly_feature = PolynomialFeatures(degree=2) # 차수=2
x_poly = poly_feature.fit_transform(x.reshape(-1,1))
```

```python
model = LinearRegression()
model.fit(x_poly,y)

y_pred = model.predict(x_poly)
```

```python
plt.plot(x,y_pred)
plt.scatter(x,y)
plt.xlabel('day')
plt.ylabel('number of infection')
print('R_squared : ', r2_score(y,y_pred))
print('beta0 :',model.intercept_)
print('beta1 :',model.coef_[1])
print('beta2 :',model.coef_[2])
print('y = {} x^2 + {} x + {}'.format(model.coef_[2],model.coef_[1],model.intercept_))
```

실제로 중국내 총 확진자의 수는 이차함수의 곡선과 일치하게 증가하는 것 처럼보인다.  
학생의 주장대로 초기 포인트들의 $R^2$ 값은 상당히 높다(99.6%)
<br>

하지만 문제는 두가지.
* 인구성장/경제성장 및 질병확산 등 초기 기하급수적으로 증가하는 데이터들은 대체로 모두 $R^2$ 값이 높다.  
    -> 초기의 높은 $R^2$ 값으로 모델의 정확도를 증명하기에는 섣부르다
* 이차함수의 그래프 특성상 한계없이 발산하는 그래프이다.  
    -> 제한없는 발산은 불가능하다.

```python
y_ = c.values
t = np.arange(len(y_))
t_poly = poly_feature.fit_transform(t.reshape(-1,1))
plt.scatter(t,y_, label='17~ data')
plt.plot(t,model.predict(t_poly), label='predict', c='orange')
plt.scatter(x,y, label = '~17 data')
print('R_squared : ', r2_score(y_,model.predict(t_poly)))
plt.xlabel('day')
plt.ylabel('number of infection')
plt.legend()
```

## 2.logistic growth model(SI model)
<br>

중국의 확진자 수 조작이라는 주장으로 이슈가 적잖이 달아올랐을 무렵 이를 반박하는 기사가 올라왔다.

### " 코로나 2월말 99%도달 3월초 종료"
![사진2](https://pds.joins.com/news/component/htmlphoto_mmdata/202002/13/be979c7a-563c-4ee4-888a-12581465e980.jpg)
<br>

제한없이 exponential하게 증가하는 모형의 한계는 뚜렸하다. 사회과학적인 입장에서 초기의 가파른 성장과 높은 $R^2$는 당연한 것이라는 주장과 함께 또 다른 모델이 등장했다.
<br>

분석을 하신 교수님께서 제시한 모델은 여러 질병모델의 모태가 되는 **susceptible infectious(SI)** 이다. 이 모델은 우리가 흔히 잘 아는 **logistic growth** 모델과 같은 모양을 가지고 있다. 위의 그림에서 보여주듯이 초기에는 가파른 성장을 하지만 어느 시점이 되면 그 성장세가 둔화되면서 특정 값으로 수렴하는 경향을 보여준다. 무제한적인 증가가 아니라 증가치의 한계를 설정한 것이다.  그 식을 보면 다음과 같다. 

$$
\begin{align} \frac {dy} {dt} = & \frac {\beta} {K} Sy,\\\\
    = & \frac {\beta} {K}(K-y)y, \\\\
    = & \beta y - \frac {\beta} {K} y^2
    \end{align}
$$

* K : carrying capacity
* y : cumulative infection
* &beta;: infection rate
* t : time point
* S + y = N
<br>

이차함수 그래프나, 지수함수의 형태를 띄는 모델과 다른점은 이 모델은 어떤 **한계point** 을 가지고 있다는 점이다.무제한적으로 발산하던 이전 모형과 달리 초기에는 빠른 속도로 증가하다 그 증가속도가 줄어들면서 수렴하게 된다.인구증가 또는 경제발전과 같은 사회현상을 잘 설명할 뿐 아니라 생태계 내 종의 확산과 같은 자연현상도 잘 설명한다. **carrying capacity**라고 불리는 K 값은 모형의 인구 증가의 한계치를 의미한다. 질병확산모델로 대입을 시키면 **감염자수의 최대치** 라고 생각할 수 있다. &beta;는 감염률을 의미하고 지수증가 형태의 모델과는 큰 차이가 없다.<br>


$$
\begin{align} \frac {dy} {dt} = & \frac {\beta} y \frac {(K-y)} {K} \\\\
    = & {\beta} y {(1- \frac {y} {K})} \\\\
    \approx &  \beta y
    \end{align}
$$
<br>수식의 우항을 살펴보면 (1-y/K) 부분이 곱해져있다. 이 부분이 없으면 일반적인 지수함수 그래프와 동일한 형태이다. 의미적으로 살펴보면 y/K는 감염자의 한계치 대비 해당 t시점에서의 누적 감염자 수 이다. y는 점점 누적되며 K와 종래에는 같아진다. 하지만 초기에는 y값이 아주 작다. 따라서 t가 작은 초기에는 y/K는 0과 가까운 아주 작은 값이라 예상할수 있다. 따라서 초기의 &beta; 는 다음과 같이 표현될 수 있다.<br>

$$
\frac {dy} {dt*y} = \beta
$$

<br>초기의 감염자의 추이는 지수함수처럼 기하급수적으로 늘어난다. 초기 10일까지의 감염자 추이를 logistic모형과 지수모형으로 비교해보자.

```python
def exp(t,r):
    y =c[0]*np.exp(r*t) 
    return y

def logit(t,r,k):
    y =(c[0]*k*np.exp(r*t))/((k-c[0])+c[0]*np.exp(r*t)) 
    return y

t= np.arange(10)
plt.plot(t,exp(t,0.2), label='exponential')
plt.plot(t,logit(t,0.2,100000), label='logistic')
plt.title('increasing tendency in early phase')
plt.xlabel('day')
plt.ylabel('infection')
plt.legend()
```

상당히 유사한 증가추이를 부여준다. 하지만 이는 초기 상태의 경향성을 보여줄뿐 전체 감염자의 추이를 설명할 수 없다.<br>

다시 돌아가서, 교수님께서 주장하신 logistic 모형을 살펴보자.
기본적으로 이 모형을 사용하는 경우는 한계치인 'K'값에 대한 정보를 가지고 있고 대입해서 사용한다.대개 초기 몇 포인트를 통해서 'K'를 추정하여 사용하거나, 해당 도메인 지식을 이용하여 K 를 추정한 후 사용 하는데 초기의 감염추이를 통해 K를 구하는 경우는 그 정확도는 신뢰할만 하다고 생각하지는 않고 어떠한 전염병적 역학에 대한 지식이 없기 때문에 K를 선택하는 것은 쉽지 않다.?. <br>

정확하게 교수님께서 어떤 방식으로 분석을 하셨는지 모르기 때문에, 사용하셨던 logistic 모형을 기반으로 개인적인 추론을 통해서 분석을 해보았다. <br>


### **2.1 모델 추정**
#### **2.1.1 모델**
로지스틱 모델은 다음과 같다.<br><br>

$$
y_{t} = \frac {K*exp(r*t)} {(K-p0) + exp(r*t))}
$$
<br>
* **input** : t, K, r, p0
* **output** : y_t
<br><br>

모델은 **'최적화'** 함수를 이용하여 추정할 것이다. 현재 input 변수들에 대한 정보를 모른다. 그렇기에 관측치를 가장 잘설명해주는 input 값을 최적화 시켜가며 구할 것이다. 
> *로지스틱 모형과 현재의 관측데이터를 가장 높은 정확도로 설명해주는 변수들을 최적화 한 후 모델에 사용한다.*

<br>

*  변수 설명
  * **t_f** : 실제 최초 확진자 발생일 (**target**)
  * **t_obs** : 최초 데이터 보고일(1/22일)
  * **t_lag** : t_obs - t_f (t_obs가 최초 확진자 발생후 몇일후 인지)
  * **y_t** : t 시점에서 누적 확진자 수
  * **y_pred** : 모형으로 예측한 확진자 수
  * **p0** : t_o 에 발생한 감염자수(실제 최초 감염자 수)
  * **K** : carrying capacity(**target**)
  * **r** : 감염률(**taret**)
  
*  는  최초 확진자 발생후 몇일후의 데이터 인지 말해준다.
*  p0 은 10명 이다. (가정)
*  서로 다른 't_lag' 에 대해서 K,r 을 최적화한 후, y(데이터) 와 로지스틱 모형으로 피팅한 추정치 간의 metric 성능을 가장 높게 만들어주는 't_lag, K, r '을 선택
*  metric은 custom loss 함수를 사용한다.
<br><br>

*p0값에 따라서 모델이 크게 좌우되지 않았다. 경험적으로 설정한 값이며 개연성도 가지고 있다고 생각한다. <br><br>

#### **2.1.2 단계**
분석은 다음과 같이 진행된다.<br>

1. t_lag  = i
2. t_lag 과 p0를 만족하는 K,r을 'curve fitting'을 통해서 최적화 한다.
3. 최적화한 값을 대입하여 모델 예측값으 만든다
4. 예측값과 실제 관측값 간의 loss를 구한다
5. i<- i+1
6. 1-5 반복
7. loss를 가장 작게하는 t_lag을 선택하고, 그때의 K,r 최적화 값을 얻는다.
<br><br>

#### **2.1.3 loss함수**
custo loss 는 다음식으로 나타내었다. <br><br>

$$
Loss =  [(y_{true}[0] - y_{pred}[0])^2 + \sum_{t=1}(y_{true}[i] - y_{pred}[i])]/len(y)
$$
<br>
MSE 와 MAE의 혼합형태로 첫번재 관측치에 가중치를 주었다. 관측 데이터를 가장 잘 설명하는것도 중요하지만 첫번째 관측치와 예측치 간의 오차가 적어야 이전의 t_f 시점의 특정과 현재 값의 신빙성이 생길것이라고 생각했다.




```python
from scipy.optimize import curve_fit
from sklearn.metrics import mean_absolute_error

# 로지스틱 모델 함수

p0 = 10 # 최초 감염자수 가정

def model(t,r,k):
    y =(k*np.exp(r*t))/((k-p0)+np.exp(r*t)) 
    return y

# loss 함수
def custom_loss(y_true,y_pred):
  loss = (np.square(y_true[0]-y_pred[0]) + np.sum(np.abs(y_true[1:] - y_pred[1:])))/len(y_true)
  return loss


mae_t = []
c_loss = []
for i in range(60): # t_lag 을 0부터 60까지
  t = np.arange(i,i+17)
  y = c.values[:17]
  popt, pcov = curve_fit(model,xdata=t ,ydata=y ,bounds=(0, [1, 1e7]))
  mae_t.append(mean_absolute_error(y, model(t,*popt)))
  c_loss.append(custom_loss(y, model(t,*popt)))
```

```python
df = pd.DataFrame({'mae':mae_t,'loss':c_loss})
df.loss.plot(label="loss", legend=True)
df.mae.plot(secondary_y=True, label="mae", legend=True)

print('minimum mae at :',np.argmin(mae_t))
print('minimu loss at :', np.argmin(c_loss))
```

*t_lag*가 18일때 loss가 가장 낮다. 즉 첫번째 관측치는 최초 발생이후 19일째 되는날일 때 모델과의 정확도가 가장 높았다. <br>
t_lag이 18일때의 최적화 K,r값을 구해보자

```python
# 최적화 K, r
t = np.arange(18,18+17)
y = c.values[:17]
popt, pcov = curve_fit(model,xdata=t ,ydata=y ,bounds=(0, [1, 1e7]))

t_ = np.linspace(0,50,100)
plt.figure(figsize=(8,5))
plt.plot(t_, model(t_, *popt), 'r-', label='prediction')
plt.plot(t, model(t, *popt), 'b-', 
         label='fit: r=%5.3f, k=%5.3f' % tuple(popt))
plt.scatter(t,y, label='data_before')
plt.scatter(np.arange(35,45),c.values[17:], label='data_after')
plt.axvline(x=34, label='prediction_point', c='red', ls = '--')
plt.xlabel('days')
plt.ylabel('total infection')
plt.legend()
```

### **2.2 평가**
첫번째 관측치가 전염병 발발이후 19일째의 데이터라고 추정했다. 데이터를 가정하고 그려본 예측 값이다. 현재 데이터에서 최적의 'K'와 'r'을 구할 수 있다. 43000명을 *'day 35'* 까지의 값을 기준으로 예측한 값은 이후 관측치를 비슷하게 예측하다가 *'day 40'* 이후의 값에 잘 대응하지 못한다.<br>

## **3.logistic with Bayesian**

로지스틱 모델을 베이지안 방법론으로 분석 해보면 어떨까? .<br>

기본적으로 베이지안 모델이 가지는 장점이라고 하면 **불확실성에 대한 계량화**가 가능하다는 점이다. 모델이 결과로 내놓는 값들이 얼마나 신뢰도가 있는지에 대해서는 기존 모델들은 어떠한 정보도 주지 못한다. 하지만 베이지안 모델이라면 결과값들이 얼마나 신뢰도가 높은지 사후분포를 통해 알 수있다.<br>

또한 모수 추정에 있어서 분석가의 **주관적인 사전분포**를 정함으로써 샘플수가 적은경우나 편향된 샘플인 경우 효과적으로 대처할 수 있는 장점이 있다.

기존 로지스틱에 추가/고려 되는 점이라고 하면
* 모수에 해서 사전 분포를 가정한다.
* 샘플링이 되면서 기존의 prior이 업데이트 된다.
* 불확실성을 알수 있다.

이정도 일것이다.
<br><br>

### **3.1 basic model.**

동일한 로지스틱 모델을 베이지안 방법을 이용하여 분석해보자.<br>**t_lag** 은 최적화를 통해 앞서 구했던 값으 그대로 사용하여 분석을 시작해보겠다..<br><br>

### **3.1.1 non-informative prior.**
모수에 대해서 사전정보가 부족한 경우 정보량이 많지 않은 prior분포를 사용하는것이 분석의 시작일 것이다.


```python
# 로지스틱 모델
def logit_model(t,r, K, p0):
    y =(K*np.exp(r*t))/((K-p0)+np.exp(r*t)) 
    return y

p0 = 10 # 최초감염자 수
t = np.arange(18,18+17) # day 18~35 까지 17개의 관측 시점
y = c[:17].values # 17개의 관측치
```

```python
# 1번 모델 : non-informative prior
with pm.Model() as model:
  
    K = pm.Uniform('K',lower=0 , upper=500000)    
    r = pm.Beta('r',alpha=2,beta=2)   
    
    theta = (logit_model(t,r,K,p0))
    sig = pm.HalfCauchy('sig', 1000)
    
    y_obs = pm.Normal('y_obs', mu=theta, sigma=sig , observed=y)
    
    
pm.model_to_graphviz(model)
```

모델이 가지고 있는 모수는 총 3개이다. 정보량이 적은 non-informative를 prior을 사용하려고한다. 사전분포의 정보에 대해서 간단히 설명하면 다음과 같다. 
* K (carrying capacity) : 누적감염 수렴 지점. 정보량이 많지 않은 flat한 형태의 prior인 'uniform'을 사용
* r (infection rate) : 감염률. 감염률의 경우 값이 (0,1) 이므로 beta분포를 가정하면 좋은 prior이 될것.
* sig (standard deviation) : likelihood에 사용될 편차. 편차값이므로 0보다 커야하기에 halfcaucy 분포를 사용.

```python
with model:
    trace =pm.sample(3000, tune=3000)
```

```python
import arviz as az
az.plot_trace(trace)
```

```python
pm.summary(trace)
```

초기 17개의 포인트를 이용하여 사후분포에서 샘플링을 한 결과, 
* K는 44000 지점 근방에 수렴한다
* r은 0.35 지점 근방에서 수렴한다
<br>

이를 바탕으로 모델을 그래프로 나타내면 다음과 같다.

```python
t_ = np.arange(18,18+len(c))
plt.plot(t_, logit_model(t_, 0.35, 43500, p0), label='predict')
plt.plot(t,logit_model(t,0.35,43500, p0))
plt.scatter(t_,c, label='35~ data')
plt.scatter(t,c[:17], label='~35 data')
plt.axvline(x=34, c='red', ls ='--')
plt.xlabel('days')
plt.ylabel('number of infection')
plt.title('bayesian inference')
plt.legend()
```

사후분포에서의 평균값을 점추정치로 사용하였다. 추정한 값을 이용해 로지스틱 모델을 보면 예측당시 관측된 데이터들과 예측선은 상당히 일치하는 경향을 보여주지만 이 후 관측치에서는 잘 대처하지 못한다. <br>

### **3.1.3 updating prior.**

베이지안의 장점이라고 하면 정보를 업데이트 할 수 있다는 점이다. 내가 가진 prior과 관측되는 데이터를 통해 우리는 계속해서 prior을 업데이트 할 수 있다.  만약 추가로 데이터가 관측이 되었다면 posterior은 새로운 데이터에 대한 prior로 사용될 것이다. <br><br>

model2에서 얻은 posterior을 이용하여 35일 이후 얻은 데이터들을 추론하는데 사용해보자.

```python
t = np.arange(18+17,18+27)
plt.scatter(t,c[17:])
plt.title('n_infection after day35')
```

*'day 35'*이 이후 데이터는 모델과는 차이가 크다. *'day 40'*일 까지의 데이터는 모델에 잘 피팅 되어 나타나다가 갑자기 확진자가 급증했다.<br>
바뀐 통계산정방식 으로 데이터가 바뀌었다. '모델2' 에서 얻은 posterior을 업데이트된 나의 prior로 간주하고 *'day 40'* 일까지의 데이터를 추가로 분석해보자.

```python
# posterior 정보를 사용하기 위한 함수.

def from_posterior(param, samples):
    smin, smax = np.min(samples), np.max(samples)
    width = smax - smin
    x = np.linspace(smin, smax, 100)
    y = stats.gaussian_kde(samples)(x)

    x = np.concatenate([[x[0] - 3 * width], x, [x[-1] + 3 * width]])
    y = np.concatenate([[0], y, [0]])
    return Interpolated(param, x, y)
```

```python
y = c[17:22].values # 35~40일 데이터
t = np.arange(18+17,18+22) # time point
p0 = 10

with pm.Model() as model3:
        
  # 모수였던 K,r 에대해서 posterior을 prior로 사용
  K = from_posterior('K', trace2['K']) 
  r = from_posterior('r', trace2['r']) 
  
  sig = pm.HalfNormal('sig', 1000)
  theta =(logit_model(t,r,K,p0)) 

  y = pm.Normal('y_obs', mu=theta, sd=sig , observed=y)
```

```python
with model3:
  trace3 = pm.sample(3000, tune=10000)
```

### **3.2 switching model.**
 35일 이후 데이터가 말해주듯이 데이터 집계 방식이 변해서 데이터의 트렌드가 변해버렸다. `이를 미리 예측하기는 불가능하다.`  
 그러면 사후적으로 , 이런 바뀐 양상을 포함하는 모델을 만들어보자.<br><br>

### **3.2.1 switching model 1**
여기서 가정할 것은 다음과 같다.
* K 값의 갑작스러운 변화. 특정시점에서 K1 -> K2로 변한다.
<br>

'switch 함수'를 새로이 도입했다. 이 함수의 역할을 말 그대로 'K'가 바뀌는 시점을 기준으로 K의 값을 변화시켜주는 함수이다.




```python
y = c.values # 전체 데이터 
p0 =10
t = np.arange(18,18+len(c))

# switch function.
def switch_func(L, x0, k=100, t=np.linspace(18., 18+len(c), len(c))):
  return L/(1+np.exp(-k*(t-x0)))  

plt.plot(t, switch_func(100,30,k=100), label='func_1, k=100')
plt.plot(t, switch_func(100,30,k=0.5), label='func_2, k=0.5')
plt.legend()
```

그래프에서 보여주듯 이 함수는 0 에서 특정값으로 바꾸어주는 함수이다. 바꾸는 정도를 'k'를 통해서 조절할 수 있다. 갑작스러운 변화인지 점진적인 변화인지 그 정보를 줄 수 있다.

```python
## 모델4 : switch model1 - K

with pm.Model() as model4:
    
    switchpoint = pm.DiscreteUniform('switchpoint', lower=t.min(), upper=t.max())  # 변화는 시점.

    early_K = pm.Normal('early_K', mu=50000, sd=5000) # K1
    late_K = pm.Uniform('late_K', lower=0, upper=150000) # K2
    K = early_K + switch_func(late_K, switchpoint) # ealry_K + late_K

    r = pm.Beta('r',alpha=2, beta=2)           
    sig = pm.Uniform('sig', lower=0, upper=10000)
    
    theta =  (logit_model(t,r,K,p0))
    y_obs = pm.Normal('y_obs', mu=theta, sd=sig , observed=y)
    
    
pm.model_to_graphviz(model4)
```

```python
with model4:
    trace4 = pm.sample(2000, tune=10000)
```

```python
import arviz as az
az.plot_trace(trace4)
```

```python
pm.summary(trace4)
```

꽤 재밌는 점은 갑작스러운 변화가 발생한 *'day 40'* 지점에서 '모델4' 역시 100%의 강한 확신으로 변화를 감지했다. 어느정도로 확실한 변화인지를 수치적으로 제공하는 점은 베이지안의 장점이라고 할 수 있다.<br>

```python
# 300개의 샘플 그래프.

t= np.arange(18,18+len(c))
for i in range(300):
  K = trace4[1950-i]['early_K'] + switch_func(trace4[1950-i]['late_K'], trace4[1950-i]['switchpoint'],k=1)
  theta =  logit_model(t,trace4[1950-i]['r'],K,p0)
  plt.plot(t,theta, alpha=0.01, color='red')
plt.scatter(np.arange(18,18+len(c)),c, color='black')
plt.xlabel('day')
plt.ylabel('number of infection')
plt.title('trend of infection')
```

MC 샘플링을 통해서 얻은 샘플 300개에 대한 그래프를 그려보았다. 불확실성이 크지 않은 꽤 괜찮은 추론을 보여준다.

### **3.2.2 switching model 2**
'모델4' 에서는 K의 갑작스러운 변화 point를 통해서 모델을 만들었다. 이번에는 감염률(r) 에 대해서도 마찬가지의 고려를 해보자.<br>
모델은 감염률을 고정된 상수로 간주하지만 실제로 생각해보면 `감염률 이라는 것은 상당히 가변적이다.` 초기 전염병이 퍼지기 시작하는 지점에서는 전염에 대한 사회적인 인식이라던지 심각성에 대해서 충분히 논의 되지 못하기에 사람간의 접촉이나 감염을 일으키는 요인들에 대해 잘 대처하지못해 감염률이 상대적으로 높을 것이라 생각할 수 있다. 하지만 그 이후 전염병에 대한 사회적 통제 또는 인식의 변화로 감염을 일으키는 요인들을 다양한 방법으로 통제하게 될것이고 결과적으로는 감염률이 적어질것이다.<br>
K와 마찬가지로 다음을 고려한다.
* 특정 시점 에서 감염률(r)의 변화: r1->r2

```python
## 모델5 : switch model 2

with pm.Model() as model5:
    
    early_K = pm.Normal('early_K', mu=50000, sd=5000)
    late_K = pm.Uniform('late_K', lower=0, upper=150000)
    K = early_K + switch_func(late_K, 40) # K 변화 포인트는 이미 구했기 때문에 상수로 부여.

    switchpoint_r = pm.DiscreteUniform('switchpoint_r', lower=t.min(), upper=t.max()) 
    early_r = pm.Beta('early_r',alpha=2, beta=2)
    late_r = pm.Uniform('late_r',lower=-0.5, upper=0.5)
    r = early_r + switch_func(late_r, switchpoint_r, k=0.5)     

    sig = pm.Uniform('sig', lower=0, upper=10000)
    
    theta =  (logit_model(t,r,K,p0))
    y_obs = pm.Normal('y_obs', mu=theta, sd=sig , observed=y)
    
    
pm.model_to_graphviz(model5)
```

```python
with model5:
  trace5 = pm.sample(2000, tune=10000)
```

```python
az.plot_trace(trace5)
```

```python
pm.summary(trace5)
```

'모델5'는 *'day 36'* 전후로 감염률이 변하고있다고 모델을 평가했다. '모델4'와 다른점은 'switch 함수'의 'k'값이 아주 높아 급격히 변화했던 '모델4'에 비해 '모델5'에는 낮은 'k' 값을 주어 서서히 변화하는 감염률을 가정했다. 기대했던 부분은 사회전반적인 부분에서의 점진적인 변화였지만 그것 보다 갑작스런 데이터의 변화에 더욱 민감히 반응했다.(모델4 와 같은 k 일때는 동일하게 100%확신으로 *'day 40'*을 변화 지점으로 지목했음)

**모델4 vs 모델5**


```python
# 300개의 샘플 그래프.
t= np.arange(18,18+len(c))
for i in range(500):
  K = trace4[1950-i]['early_K'] + switch_func(trace4[1950-i]['late_K'], trace4[1950-i]['switchpoint'],k=1)
  theta =  logit_model(t,trace4[1950-i]['r'],K,p0)
  plt.plot(t,theta, alpha=0.01, color='red') 
  K2 = trace5[1950-i]['early_K'] + switch_func(trace5[1950-i]['late_K'], 40,k=1)
  r2 = trace5[1950-i]['early_r'] + switch_func(trace5[1950-i]['late_r'], trace5[1950-i]['switchpoint_r'],k=1)
  theta2 =  logit_model(t,r2,K2,p0)
  plt.plot(t,theta2, alpha=0.01, color='blue')
plt.scatter(np.arange(18,18+len(c)),c, color='black')
plt.xlabel('day')
plt.ylabel('number of infection')
plt.title('trend of infection')
plt.legend()
```

베이지안 추론은 대개 샘플링의 방법을 통해 이루어진다. posterior 분포를 직접 구하는 것이 어렵기 때문에 간접적으로 샘플링의 방식을 택한다. MCMC 샘플링의 경우 수렴에 대한 가정이 충족되어야 한다. r_hat 은 '1' 에 가까운 값이면 (구체적으로 1.1미만) 수렴의 조건을 충족했다고 간주한다.

ppc 그래프에서 보여주듯 posterior에서 샘플링한 데이터와 시제 관측치와 아주 유사하게 진행함을 알 수 있다. 예측에 대한 내용은 아니지만, 현재 데이터를 잘 설명해주는 모델을 만들었다고 생각할 수 있다.
