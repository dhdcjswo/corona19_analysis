### 1.2  검사횟수와 확진자

그렇다면 검사횟수가 늘어나면 확진자수는 어떻게 될까? 
예를 들어보자. 논의를 간단하게 하기 위해서 감염병이 아니라 고정된 환자라고 가정한다.(확산이 진행되지 않는 상황) 어떤 유증상자/의심환자 10000명이 있는 사회가 있다. 의심환자들 모두를 검사하려고 한다. 하루 검사능력이 100명인 경우와 500명인 경우 확진자 수는 어떨까.10000명중 무작위로 검사한다고 하자.

* **n(의심환자)** = 10000
* **a(감염자)** =  300
* **b(비감염자)** = 9700


```
'''
* 0~299 : 감염자
* 299~9999 : 비감염자
'''

n_100 = np.arange(10000) 
n_500 = np.arange(10000) 

inf_trend_100 = [0]
inf_trend_500 = [0]
for i in range(20):  
  test100 = np.random.choice(list(n_100),100, replace=False) # 검사능령 100
  test500 = np.random.choice(list(n_500),500, replace=False) # 검사능력 500
  n_infection_100 = np.sum(test100<100)
  n_infection_500 = np.sum(test500<100)
  inf_trend_100.append(n_infection_100 + inf_trend_100[-1])
  inf_trend_500.append(n_infection_500 + inf_trend_500[-1])
  n_100 = set(n_100) - set(test100)
  n_500 = set(n_500) - set(test500)
```

```
plt.plot(inf_trend_100, label='test100')
plt.plot(inf_trend_500, label='test500')

plt.xlabel('days')
plt.ylabel('n_infection')
plt.legend()
```

검사능력이 500명일때 확진자수가 많다. 검사를 많이 하면 검사에 포함되는 감염자수가 늘어나기때문에 확진자가 늘어나는 것은 당연한 결과이다.<br><br>
위의 예시는 고정된 환자에 대한 내용었지만 환자가 고정되어 있지 않고 증가하는 상황이라도 마찬가지로 검사횟수와 확진자는 비례관계를 가진다.고정된 경우에도 확진이 늘어나는데 환자수가 늘어난다면 더더군다나 확진자수는 증가할 것이다.<br>

 **'검사 횟수가 많으면 확진도 많아진다'**.<br>


*but*, 우리가 근본적으로 알고싶은 것은 사회 내 퍼지고 있는 전염의 양상이지 단순이 절대적인 숫자의 증가를 알고싶은 것이 아니다. 환자가 고정되어있는 경우에도 확진자는 늘어난다. 단순히 증가하는 확진자수로 확산을 평가하는것은 무리가 있다.

### 1.3 검사량 대비 확진자수

검사량 대비 확진자수의 추이는 무엇을 이야기 해줄 수 있을까. t 시점에서 우리는 모르지만 전체 감염자를 $I_t$ 라고하자. $I_t$는 두가지 경우로 나눌수 있다.
* 검사o : $\alpha$*$I_t$
* 검사x : $(1-\alpha)$*$I_t$ 
* $\alpha$ : 검사받을 비율 = 전체감염자 중 확진받은 비율
<br>

검사를 받지 않은 이유는 아마 '잠복기' 또는 무증상으로 병의 감염을 인지를 못하는 경우이거나 유사 증상이 발생했지만 검사를 대기하는 경우일 것이다.<br>

현재 우리가 인지할 수 있는 감염자의 숫자는 실제 감염자들중에서 검사를 받은 사람들, $\alpha$*$I_t$ 이다. $\alpha$가 1에 가까우면 현재의 확진자와 실제 감염자의 수가 비슷하다고 할 수 있고, 반대로 $\alpha$가 0에 가까워지면 실제 감염자를 과소평가 하고 있다고 해석할 수 있다.<br>

$\alpha$ 는 고정된 값이 아니다. 감염확산 초~중기에는 어느정도의 비율을 유지하다가 감염이 종식되는 시기에는 점차 1에 근접하게 될것이고 결국 수렴하는 값일 것이다.

우리는 현시점에서 실제 감염자수를 모른다. 그럼에도 불구하고 우리는 추세를 분석하고 상황을 판단하려고 노력한다. 이는 현재 확진자의 증가 추이가 실제 감염의 증가추이를 '설명'해줄 것이라는 믿음에서 비롯하고, 이 믿음은 신념 따위가 아닌 통계적인 근거에 의한 믿음이다.<br>  
이 믿음은 아마도, 현재의 확진자의 추이가 실제 감염자의 추이를 설명해줄 수 있는 대표성을 띄고 있다는 믿음일 것이고 이는 곧 **샘플링** 문제와도 연결이 될것이다.

검사를 받는 경우를 생각해보자. 검사를 받는 사람들의 경우는 다음과 같다.
* $T$ : 전체 검사자
* $\alpha$*$I_t$ : 실제 감염자
* $T$ - $\alpha$*$I_t$ : 유사증상/의심환자 중 감염이 아닌 경우

아까 그래프를 다시보자 





```
plt.plot(inf_trend_100, label='test100')
plt.plot(inf_trend_500, label='test500')

plt.xlabel('days')
plt.ylabel('n_infection')
plt.legend()
```

```
'''
* 0~299 : 감염자
* 299~9999 : 비감염자
'''

diff_100 = []
diff_500 = []
for i in range(1,len(inf_trend_100)):
  diff_100.append(inf_trend_100[i] - inf_trend_100[i-1])
  diff_500.append(inf_trend_500[i] - inf_trend_500[i-1])

plt.plot(np.array(diff_100)/100)
plt.plot(np.array(diff_500)/500)
```

위 그래프는 일일 검사자 대비 확진자의 비율을 나타낸 것이다. 확진자의 절대 수치에서는 차이가 났지만 비율 그래프에서는 비슷한 양상을 보여준다. 사회 내 감염자들의 추이를 설명할 수 있는 지표가 될 수 있는 그래프이다. 사회 내에서 확진자들이 차지하는 비율이 늘어난다면 확진자/검사자 의 비율은 높아질 것이다. 위의 예시처럼 확산이 더이상 이루어지지 않는 경우에는 유지 될것이다.<br><br>

### 1.5 검사비율($\alpha$)

검사비율($\alpha$)은 왜 중요할까?

$\alpha$이 높지 않다면 추가적인 전염확산에 취약할 수 밖에 없다. 확진자들은 격리를 통해서 관리가 되지만 아직 검사를 받지 않은 감염자들에 대해서는 조치가 어렵다. <br>

$\alpha$이 높다는 말은 사회가 감염자들을 잘 통제하고 있는 상황이라고 봐도 무방하다. 이 비율이 높다면 선제적인 조치를 통해 전염의 확산을 빠르게 막을수 있고, 현재 상황을 정확히 인식하고 조치를 취할 수 있다. **의견1** 의 주장으로 간주했던 이 비율은 현재 상황에 대한 정보라기보다 통제와 조치를 통한 앞으로의 확산에 영향을 주는 요인으로 이해할 수 있다. 
또한 위에서 기술했듯이 $\alpha$ 값의 특성상 감염이 종식되어 간다면 1로 수렴하는 값이다. $\alpha$이 높으면 전염병이 끝나가고 있음을 말해주는 지표라고도 할 수 있다.<br><br>

**하지만 $\alpha$값은 우리가 전염병이 종식되기전까지는 알수 없는 값이다.**
